```python
#Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
max_score = max(employees.values())

# Find all employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Format the output
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
# Search Query Keyword Analysis
import string

# Take the search query from the user
query = input("Enter search query: ")

# Convert the entire query to lowercase so that the counting is case-insensitive
query = query.lower()

# Remove all common punctuation marks to avoid treating words like 'phone,' and 'phone' differently
query = query.translate(str.maketrans('', '', string.punctuation))

# Split the sentence into individual keywords
words = query.split()

# Create an empty dictionary to store the frequency of each keyword
frequency = {}

# Count how many times each keyword appears in the query
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Keep only those keywords that are searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

# Display the final result
if result:
    print(result)
else:
    print("No keywords searched more than once")
```

```python
#Sensor Data Validation
# Take sensor readings as input from the user and convert them into a list of integers
sensor_readings = list(map(int, input("Enter sensor readings separated by space: ").replace(',', ' ').split()))

# Create an empty list to store the valid (hour, reading) pairs
valid_readings = []

# Traverse through the list using index and value together
for hour, value in enumerate(sensor_readings):

    # Check whether the reading is even, since only even readings are considered valid
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Display the valid sensor readings in the required format
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis
# Take multiple email IDs as input from the user
emails = input("Enter email IDs separated by space: ").split()

# Dictionary to store the count of each domain
domain_count = {}

# Extract the domain from each email and update its count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Total number of email users
total_users = len(emails)

# Display the percentage usage of each domain
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Sales Spike Detection
# Take daily sales values as input from the user
sales = list(map(int, input("Enter daily sales separated by space: ").replace(',', ' ').split()))

# Calculate the average sales value
average_sales = sum(sales) / len(sales)

# Define the spike threshold as 30% above the average
threshold = average_sales * 1.30

# Identify and display the days where sales exceeded the spike threshold
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Duplicate User ID Detection
#Take user IDs as input from the user
user_ids = input("Enter user IDs separated by space: ").split()

# Create a dictionary to store the frequency of each user ID
frequency = {}

# Count how many times each user ID appears
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

# Display only the user IDs that appear more than once
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
