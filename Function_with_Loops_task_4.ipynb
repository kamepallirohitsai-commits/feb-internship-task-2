```python
#Mobile Recharge Validation System
def validate_recharge():
    valid_plans = {199, 299, 399, 599}

    while True:
        try:
            amount_str = input("Enter recharge amount (₹): ")
            amount = float(amount_str)

            plan_str = input(f"Enter recharge plan (valid plans: {', '.join(map(str, sorted(valid_plans)))}): ")
            plan = int(plan_str)

            # Rule 1: Recharge amount must be >= ₹50
            if amount < 50:
                print("Invalid recharge amount. Amount must be ₹50 or greater.")
                continue  # Allow retry

            # Rule 2: Valid plans
            if plan not in valid_plans:
                print(f"Invalid plan. Valid plans are: {', '.join(map(str, sorted(valid_plans)))}.")
                continue  # Allow retry

            print("\nRecharge request validated successfully!")
            print(f"Amount: ₹{amount:.2f}")
            print(f"Plan: ₹{plan}")
            break  # Exit loop if valid

        except ValueError:
            print("Invalid input. Please enter a number for amount and an integer for plan.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Call the function to test the validation system
validate_recharge()
```

```python
#Inventory Reorder Alert System
def check_inventory():
    inventory = {'ProductA': 20, 'ProductB': 10, 'ProductC': 50, 'ProductD': 5, 'ProductE': 12}
    for product, stock in inventory.items():
        if stock < 15:
            print(f"Reorder Alert: {product} (Stock: {stock})")
        else:
            print(f"Stock OK: {product} (Stock: {stock})")

check_inventory()
```

```python
#Student Result Processing System
def process_student_results(marks):
    average = sum(marks) / len(marks)
    if average >= 50:
        print('Pass')
    else:
        print('Fail')

process_student_results([60, 70, 80, 55, 65])
process_student_results([30, 45, 50, 40, 35])
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(km, peak):
    fare = 50 + 12 * km
    if peak:
        fare *= 1.25
    return fare

while True:
    km = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
    peak = peak_input == "yes"
    fare = calculate_fare(km, peak)
    print(f"Estimated Fare: ₹{fare:.2f}")
    retry = input("Do you want to calculate again? (yes/no): ").strip().lower()
    if retry != "yes":
        break
```

```python
# Employee Attendance Eligibility Checker
def check_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

records = input("Enter attendance (P/A separated by space): ").split()
result = check_eligibility(records)
print(result)
```

```python
#Password Strength Checker
def check_password_strength(password):
    if len(password) < 8:
        return "Weak"
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong"
    return "Weak"

pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
