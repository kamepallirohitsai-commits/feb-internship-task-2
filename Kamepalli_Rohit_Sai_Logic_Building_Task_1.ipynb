```python
# User Login Check

correct_username = "admin"
correct_password = "1234"

username = input("Enter username: ")
password = input("Enter password: ")

if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students  :", len(marks))
print("Students Passed :", pass_count)
print("Students Failed :", fail_count)
```

```python
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print("Original Names :", names)
print("Cleaned Names  :", cleaned_names)
```

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print(f"Message: '{msg}' → Length: {len(msg)}")

    if len(msg) > 10:
        print("⚠️ This message is longer than 10 characters")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("System Logs :", logs)
print("Total ERROR messages :", error_count)
```
