```python
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest = max(employees, key=employees.get)

print("Highest Salary:", highest, "-", employees[highest])
```

```python
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]

present = attendance.count("P")
percentage = (present / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
text = "python"

freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)
```
